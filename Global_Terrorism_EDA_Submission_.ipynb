# **Project Summary -**


This project performs exploratory data analysis (EDA) on the Global Terrorism Database to understand how terrorism patterns have changed over time, geography, attack methods, and human impact. The dataset contains incident-level records from 1970 to 2017 and includes event date, location, attack type, target category, weapon type, casualties, and related contextual attributes. Because terrorism is dynamic and region-specific, a structured EDA helps convert raw event records into decision-ready intelligence for governments, security agencies, and policy planners.

The analysis starts with data understanding and quality checks. We inspect row-column dimensions, variable types, duplicates, and missingness patterns. Core columns such as year, country, region, attack type, target type, weapon type, deaths, and injuries are standardized. Date components are combined into a single event date where possible, while invalid month/day values are corrected safely before conversion. Numeric casualty fields are cleaned and imputed conservatively for descriptive analysis. Duplicate records are validated and removed where required. Outlier handling is performed with an IQR-based approach for highly skewed casualty variables, and both capped and uncapped logic are discussed because extreme events may be statistically rare but operationally critical.

Visualization is structured using the UBM rule: univariate, bivariate, and multivariate analysis. Univariate charts establish baseline distributions such as yearly incident volume, regional concentration, attack methods, and casualty spread. Bivariate charts reveal relationships like year versus attack count, country versus event burden, attack type versus lethality, and region versus casualties. Multivariate plots then combine dimensions (for example region-year trends and weapon-target-casualty relationships) to uncover compounding risk structures. The notebook includes heatmaps and pair plots to capture correlation behavior and variable interaction in a compact way.

The central findings typically expected from this dataset include: (1) terrorism activity is not uniformly distributed across years, with clear escalation phases; (2) incident concentration is heavily regional, indicating persistent hotspots; (3) a few attack and weapon categories account for a large share of total events; and (4) casualty outcomes differ substantially across attack-target combinations, which matters for prevention prioritization.

From a business and governance perspective, this analysis supports three critical outcomes. First, risk prioritization: agencies can allocate surveillance and prevention resources to high-incidence geographies and periods. Second, tactical preparedness: understanding dominant attack-weapon-target combinations improves protection design for vulnerable target classes. Third, policy calibration: multi-year trend evidence helps evaluate whether interventions are reducing incident frequency or only shifting attack behavior.

Overall, this EDA transforms historical terrorism records into actionable intelligence. It does not replace predictive or causal modeling, but it builds the essential analytical foundation required for forecasting, threat scoring, resilience planning, and data-driven counter-terror strategy.

## 3. ***Data Wrangling***


### Data Wrangling Code


````python
# Write your code to make your dataset analysis ready.
eda_cols = ['eventid','iyear','imonth','iday','country_txt','region_txt','city','attacktype1_txt','targtype1_txt','weaptype1_txt','nkill','nwound','success']
eda_df = df[eda_cols].copy()
eda_df = eda_df.drop_duplicates()
eda_df['imonth'] = eda_df['imonth'].replace(0, 1)
eda_df['iday'] = eda_df['iday'].replace(0, 1)
eda_df['event_date'] = pd.to_datetime(eda_df['iyear'].astype(str) + '-' + eda_df['imonth'].astype(int).astype(str) + '-' + eda_df['iday'].astype(int).astype(str), errors='coerce')
for c in ['country_txt','region_txt','city','attacktype1_txt','targtype1_txt','weaptype1_txt']:
    eda_df[c] = eda_df[c].fillna('Unknown')
eda_df['nkill'] = pd.to_numeric(eda_df['nkill'], errors='coerce').fillna(0)
eda_df['nwound'] = pd.to_numeric(eda_df['nwound'], errors='coerce').fillna(0)
eda_df['casualties'] = eda_df['nkill'] + eda_df['nwound']
q1 = eda_df['casualties'].quantile(0.25)
q3 = eda_df['casualties'].quantile(0.75)
iqr = q3 - q1
upper_cap = q3 + 1.5 * iqr
eda_df['casualties_capped'] = eda_df['casualties'].clip(upper=upper_cap)
eda_df['decade'] = (eda_df['iyear'] // 10) * 10
print('Cleaned shape:', eda_df.shape)
eda_df.head()
````

### What all manipulations have you done and insights you found?




I selected core variables, removed duplicates, standardized date components, and built a valid `event_date` for time-series analysis. Missing categorical values were labeled as `Unknown` to preserve row-level records during groupby operations. Casualty fields were converted to numeric and imputed with zero for aggregation consistency.

Because casualty data is highly right-skewed, I added an IQR-capped feature (`casualties_capped`) for stable distribution plots while preserving raw casualties for impact interpretation. A decade feature was also created for long-term trend comparison.
